Q1: Scenario: Fund available for improving airport equipment, what are the top 10 airport have the most urgent need in 2020. 
Constraint: 10 airport in US with most security delay in terms of times and minutes in 2020.

```python
%%sql
select (distinct AL.airport_name), sum(OR.security_delay), avg(OR.security_delay), count(1) as 
from ontime_reporting OR 
JOIN airport as A on OR.origin_airport_key = A.airport_key
JOIN time ON ontime_reporting.crs_dep_time_key = time.key
WHERE time.year = 2020
order by count(1) -- those airport who has the most delay
limit 10;
```
